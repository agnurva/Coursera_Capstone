#### Applied Data Science Capstone - Segmenting and Clustering neighborhoods in the city of Toronto

Problem Description - Cluster the neighborhoods in Tornonto

```python
import numpy as np
import pandas as pd 
```

```python
df = pd.read_csv('TorontoPostalCodes_with_lat_long.csv')
df.head()
```

### I have decided to work with Borough with the term Toronto. I think this will be more specific to the location of interest

```python
toronto_data = df[df['Borough'].str.contains('Toronto')].reset_index(drop=True)
toronto_data.head()
```

```python
from geopy.geocoders import Nominatim 
# Let's get the geographical coordinates of Toronto Cannada
address = 'Toronto, Cannada'

geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

```python
import folium
# create map of Toronto using latitude and longitude values
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=11)

# add markers to map
for lat, lng, label in zip(toronto_data['Latitude'], toronto_data['Longitude'], toronto_data['Neighbourhood']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
