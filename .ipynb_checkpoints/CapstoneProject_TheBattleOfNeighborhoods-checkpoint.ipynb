Let us assume that M.G road is the center of Bangalore. Let us get the latitude and longitude of MG Road

Define a Foursquare user_agent to fetch the latitude and longitude using address of MG road

```python
address = 'Indiranagar, Bangalore'

geolocator = Nominatim(user_agent="foursquare_agent")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print(latitude, longitude)
```

Now let us fetch all venues within 2000 mts of Indranagar

```python
search_query = 'Hotel Bar'
radius = 2000
url = 'https://api.foursquare.com/v2/venues/search?client_id={}&client_secret={}&ll={},{}&v={}&query={}&radius={}&limit={}'.format(CLIENT_ID, CLIENT_SECRET, latitude, longitude, VERSION, search_query, radius, LIMIT)
url
```

Let us use the above url to get the list of venues around Indranagar

```python
results = requests.get(url).json()
results
```

```python
# assign relevant part of JSON to venues
venues = results['response']['venues']

# tranform venues into a dataframe
dataframe = json_normalize(venues)
dataframe.head()
```
